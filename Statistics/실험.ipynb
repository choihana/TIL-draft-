### 실험이 무엇이죠?

통계를 배우는데 갑자기 실험에 대해서 먼저 배우게 되었다.
여기서 말하는 실험은 무엇일까?

그 동안 실험이라고 하면 과학실에서 실험장비를 갖추고 온갖 장비들을 돌리는 과학적 실험을 떠올렸었다. 

여기서 말하는 실험은 비지니스, 사회/심리적 현상 등 주로 사람에 대한 실험을 이야기 한다.
그리고 이 실험은 '인과 관계를 확인하기 위한 개입' 이라고 설명된다.

### Why, 왜 실험을 하죠?

1) 내가 생각한 가설이 맞는지 확인하려고

2) 확인하고자 하는 데이터가 없어서

3) 현실에서 벌어지는 현상의 인과관계 파악을 위해

비지니스에서 예를 들어보자.

의류 이커머스에서 SNS 광고를 시작했다.
이 광고로 들어오는 고객들은 가격, 할인에 민감한 고객이다. 라는 가설을 세웠다.

이 가설을 확인하기 위해서 SNS로 들어오는 사람에게 첫번째로 노출할 콘텐츠를
'고객 리뷰', '쿠폰', '상품설명' 중 하나로 설계하여 실험을 해볼 수 있다.


### 변수 3가지

- 독립 변수: 결과에 영향을 주는 변수 (위 예시에선 첫번째로 노출할 콘텐츠)
- 종속 변수: 독립 변수에 따라 변화되는 변수 (위 예시에선 구매율, 이탈율 등)
- 외생 변수: 실험에 영향을 미치는 다양한 외부 조건들 (위 예시에선 프로모션, 요일 등)

유의미한 실험진행을 위해서는 각 변수를 정확히 정의해야한다.
특히 외생 변수를 파악하고 잘 다루는 것이 매우 중요하겠다는 생각이 든다.
현실세계에서는 어쩔 수 없이 항상 외생 변수가 존재할 것이다.
이 외생변수를 다루는 방법에는 크게 두 가지가 있다.

1) 외생변수 통제
- 외생변수를 축소하여 통제 된 환경에서 데이터를 수집한다. 

2) 외생변수 정의하고 트래킹하여 분석에 참고
- 외생변수가 어떤 것인지 정의하고 실험 도중 외생변수가 가지는 값들을 기록해둔다음 분석 시 참고한다.

예를들어 날씨가 영향을 주는 실험을 진행한다면, 외생변수를 통제하여 화창한 날에만 실험을 진행하던지 모든날에 진행하되 날씨를 기록해두고 분석에 참고할 수 있다.

이렇게 설계된 실험 결과를 정확히 해석하는데에 앞으로 배울 수학적 통계 지식이 도움이 될 것 이다!
